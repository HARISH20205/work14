```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame({'X':np.random.normal(0,2,1000),'Y':np.random.normal(2,5,1000),'Y':np.random.normal(-4,10,1000)})
df
```

```python
for col in df:
    plt.hist(df[col],density=True,alpha=0.6,bins=50)
    
```

```python
sns.set()
plt.figure(figsize=(4,3))
for col in  df:
    #plt.hist(df[col],alpha=0.5,bins=40,density=True)
    sns.kdeplot(df[col],fill=True)
```

```python
plt.figure(figsize=(4,3))

with sns.axes_style('white'):
    sns.jointplot(df)
    
    
    
```

```python
iris=sns.load_dataset('iris')
sns.pairplot(iris,hue='species')
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='species',style='species')

```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='species',style='species',kind='line',errorbar='sd')
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='species',style='species',kind='line',errorbar=None)
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='species',style='species',kind='line',errorbar=None,
            dashes=True,markers=True)






```
